# Redis explore

```python
import redis

REDIS_HOST = "localhost"
REDIS_PORT = 6379
REDIS_PASSWORD = "recsys_redis_pass"

r = redis.Redis(host=REDIS_HOST, port=REDIS_PORT, password=REDIS_PASSWORD, decode_responses=True)
r.ping()
```

## Список всех ключей

```python
keys = list(r.scan_iter("*"))
print("Keys in Redis:", keys)
```

## Получить значение по ключу

```python
import json

KEY = "popular"

raw = r.get(KEY)
items = json.loads(raw) if raw else []
print(f"Key '{KEY}' contains {len(items)} items:")
items
```
